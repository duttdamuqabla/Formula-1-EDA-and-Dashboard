```python
pd.read_sql_query("""
SELECT
    C.constructorId,
    C.name,
    C.nationality AS Constructor_Nationality,
    R.points as Team_points
FROM constructors C
JOIN constructor_results R
    ON C.constructorId = R.constructorId
""", conn)
```

```python
# combining the above two tables 
final_table = pd.read_sql_query("""
SELECT
    R.driverId,
    D.driverRef,
    D.nationality AS Driver_Nationality,
    R.constructorId,
    C.name AS Constructor_Name,
    C.nationality AS Constructor_Nationality,
    R.position,
    R.points,
    R.number
FROM results R
JOIN drivers D ON R.driverId = D.driverId
JOIN constructors C ON R.constructorId = C.constructorId
GROUP BY R.driverId, R.constructorId
""", conn)
```

```python
# Distribution Plots for numerical columns

# Select numerical columns
numerical_columns = final_table.select_dtypes(include=np.number).columns

# Plot distributions
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(final_table[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
plt.show()
```

```python
categorical_cols = ["driverRef", "Constructor_Name"]

plt.figure(figsize=(12,5))
for i, col in enumerate(categorical_cols):
    plt.subplot(1,2, i+1)
    sns.countplot(y=final_table[col], order=final_table[col].value_counts().index[:10])
    plt.title(f"count of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation = final_table[numerical_columns].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python

sns.pairplot(final_table[numerical_columns])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_constructors = final_table['Constructor_Name'].value_counts().nlargest(10)
sns.barplot(x=top_constructors.values, y=top_constructors.index, palette='viridis')
plt.title('Top 10 Most Frequent Constructors')
plt.xlabel('Number of Entries')
plt.ylabel('Constructor Name')
plt.tight_layout()
plt.show()

```
